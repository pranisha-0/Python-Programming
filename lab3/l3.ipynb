```python
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even.")
elif num  == 0:
    print(f"{num} is zero. ")
else:
    print(f"{num} is odd.")
```

```python
#calc
n1 = float(input("Enter num1: "))
n2 = float(input("Enter num2:"))
opr  = input("Enter operator: ")

if opr == "+":
    print(f"{n1} + {n2} = {n1+n2}")
elif opr == "-":
    print(f"{n1} - {n2} = {n1-n2}")
elif opr == "*":
    print(f"{n1} x {n2} = {n1*n2}")
elif opr == "/":
    if n2 != 0:
        print(f"{n1} / {n2} = {n1/n2}")
    else:
        print("Error.")
else:
    print("Invalid operator.")
```

```python
#age category
age = int(input("ENter your age: "))
if age > 0 and age < 13:
    print("You are a child.")
elif age >= 13 and age < 20:
    print("You are a teenager.")
elif age >= 20 and age < 60:
    print("You are an adult.")
elif age >= 60:
    print("You are a senior citizen.")
else:
    print("Invalid age.")
```

```python
#month --> season
mon = int(input("Enter a month number: "))
if mon == 12 or mon == 1 or mon == 2:
    print("Winter")
elif mon == 3 or mon == 4 or mon == 5:
    print("spring")
elif mon == 6 or mon == 7 or mon == 8:
    print("summer")
elif mon == 9 or mon == 10 or mon == 11:
    print("Autumn")
else:
    print("INVALID")
```
